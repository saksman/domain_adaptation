## 3.1 DeepFish experiments

```python
# Create training / testing split for domain adaptation
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy
import csv

with open("/global/u2/s/saksman/DeepFish/data/Segmentation/Segmentation/domain_adaptation/target/target.csv", "r") as f:
    data = f.read().split('\n')
    data = numpy.array(data)  #convert array to numpy type array
    x_train, x_val = train_test_split(data, test_size=0.42)  

pd.DataFrame(x_train).to_csv("/global/u2/s/saksman/DeepFish/data/Segmentation/Segmentation/domain_adaptation/target/target_train.csv", index = False)
pd.DataFrame(x_val).to_csv("/global/u2/s/saksman/DeepFish/data/Segmentation/Segmentation/domain_adaptation/target/target_val.csv", index = False)
# manually: write header, remote quotes, remote /valid and /empty strings, move files around
```
